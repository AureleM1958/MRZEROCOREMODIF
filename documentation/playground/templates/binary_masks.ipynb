```python
import MRzeroCore as mr0
import matplotlib.pyplot as plt
```

# Generate phantoms with binary masks

```python

mr0.generate_brainweb_phantoms("bw_phantoms" , "3T")
```

```python

phantom = mr0.VoxelGridPhantom.brainweb("bw_phantoms/subject04_3T.npz")

obj = phantom.interpolate(256, 256, 32).slices([16])
obj.plot()
phantom = obj.build()


# get Binary Masks from phantom
Masks = phantom.Masks

#plot binary masks:
plt.figure()
for key, value in Masks.items():
    plt.imshow(value[:, :, 0])
    plt.colorbar()
    plt.title(key)
    plt.show()
```
